```python
import os
import pandas as pd
import numpy as np
import random
import torchaudio
from datasets import load_dataset
```

```python
path_all_labels = '../data/csv_label/mutil_label'  # 所有标签的路径
path_all_wav = '../data/original_audio_data'  # 所有音频的路径


def split_labels(directory):
    train_labels = {}
    test_labels = {}
    
    train_labels['path'] = []
    train_labels['despair'] = []
    train_labels['sadness'] = []
    train_labels['pain'] = []
    train_labels['guilt'] = []
    train_labels['confuse'] = []
    train_labels['helplessness'] = []
    train_labels['resentment'] = []
    train_labels['fear'] = []
    train_labels['numbness'] = []
    train_labels['anxiety'] = []
    train_labels['grievance'] = []
    
    test_labels['path'] = []
    test_labels['despair'] = []
    test_labels['sadness'] = []
    test_labels['pain'] = []
    test_labels['guilt'] = []
    test_labels['confuse'] = []
    test_labels['helplessness'] = []
    test_labels['resentment'] = []
    test_labels['fear'] = []
    test_labels['numbness'] = []
    test_labels['anxiety'] = []
    test_labels['grievance'] = []
    
    human_ids = []
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            human_ids.append(filename)
            
    train_scale = int(0.8 * len(human_ids))
    train_ids = human_ids[:train_scale]
    test_ids = human_ids[train_scale:]

    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            filepath = os.path.join(directory, filename)
            csv_data = pd.read_csv(filepath)
            for i, row in csv_data.iterrows():
                index = row['index']
                role = row['role']
                if int(role) == 1:
                    filename_prefix = filename.split('.')[0]
                    if filename in train_ids: # 放到训练集
                        train_labels['path'].append(os.path.join(path_all_wav, filename_prefix, '{}.wav'.format(row['index'])))
                        train_labels['despair'].append(row['despair'])
                        train_labels['sadness'].append(row['sadness'])
                        train_labels['pain'].append(row['pain'])
                        train_labels['guilt'].append(row['guilt'])
                        train_labels['confuse'].append(row['confuse'])
                        train_labels['helplessness'].append(row['helplessness'])
                        train_labels['resentment'].append(row['resentment'])
                        train_labels['fear'].append(row['fear'])
                        train_labels['numbness'].append(row['numbness'])
                        train_labels['anxiety'].append(row['anxiety'])
                        train_labels['grievance'].append(row['grievance'])
                    elif filename in test_ids: # 放到测试集
                        test_labels['path'].append(os.path.join(path_all_wav, filename_prefix, '{}.wav'.format(row['index'])))
                        test_labels['despair'].append(row['despair'])
                        test_labels['sadness'].append(row['sadness'])
                        test_labels['pain'].append(row['pain'])
                        test_labels['guilt'].append(row['guilt'])
                        test_labels['confuse'].append(row['confuse'])
                        test_labels['helplessness'].append(row['helplessness'])
                        test_labels['resentment'].append(row['resentment'])
                        test_labels['fear'].append(row['fear'])
                        test_labels['numbness'].append(row['numbness'])
                        test_labels['anxiety'].append(row['anxiety'])
                        test_labels['grievance'].append(row['grievance'])
    
    pd.DataFrame(train_labels).to_csv('../data/total.csv', index=False)
    pd.DataFrame(test_labels).to_csv('../data/test.csv', index=False)
    
    return train_labels['path'],test_labels['path']

total_files,test_files = split_labels(path_all_labels)
```
